```python
# Step 1 - Reading the API
```

```python
# Importing libraries

import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# URL of the ASOS Women Black Sneakers category

url = 'https://www.asos.fr/api/product/search/v2/categories/6456?base_colour=4&channel=mobile-web&country=FR&currency=EUR&keyStoreDataversion=p1swt7e-15&lang=fr&limit=500&offset=0&rowlength=2&store=FR'
```

```python
# Request / Response

user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'
header = {'user-agent': user_agent}

response = requests.get(url,headers=header)                         
results = response.json()
results
```

```python
# Normalize

flattened_data = json_normalize(results)
#flattened_data

flattened_data1 = json_normalize(flattened_data.products[0])
flattened_data1.shape
```

```python
# Step 2 - Collecting Data from the API
```

```python
total_items = results['itemCount']
print('The total number of items is :', total_items)
```

```python
# Step 3 - Create a dataframe
```

```python
df = pd.DataFrame()

url = f'https://www.asos.fr/api/product/search/v2/categories/6456?base_colour=4&channel=mobile-web&country=FR&currency=EUR&keyStoreDataversion=p1swt7e-15&lang=fr&limit=500&offset=0&rowlength=2&store=FR'
response = requests.get(url, headers=header)
results = response.json()
flattened_data = json_normalize(results)
flattened_data1 = json_normalize(flattened_data.products[0])
flattened_data1 = flattened_data1.set_index('id')
df = df.append(flattened_data1)

df
```

```python
# WEB SCRAPPING PROJECT
# COP COPINE WEBSITE
# SCRAPPING ALL PRODUCTS URLS > NOUVELLE COLLECTION (https://www.cop-copine.com/collection-automne-hiver-2019.html)
```

```python
# Import Libraries
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
from pprint import pprint
from lxml import html
from lxml.html import fromstring
import urllib.request
from urllib.request import urlopen
```

```python
# This is the URL to scrape

url_web = 'https://www.cop-copine.com/collection-automne-hiver-2019.html'
```

```python
html = requests.get(url_web).content
html[0:2000]
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
# Links of all new products
```

```python
tag = ['a']

link_tags = [link for link in soup.find_all(tag, {'class':'product-image'})]
```

```python
links = []

for link_tag in link_tags:
    href = link_tag.get('href')
    links.append(href)

links[0:]
```

```python
# Creating a Dataframe
```

```python
df = pd.DataFrame(links,columns=['Products URL'])

df.head(20)
```

```python
# Exporting as a CSV file
```

```python
df.to_csv('webscrapping_output.csv', sep=',', index=False)
```
