````python
h1tags = soup.find_all('h1',{'class':'h3 lh-condensed'})
h1tags


# la fonction .get permet de trouver un attribut dans un tag (librairie Beautiful Soup)

names = []


for h1tag in h1tags:
    a_tag = h1tag.find('a')
    #print(a_tag)
    user_name = a_tag.get('href')[1:]
    full_name = a_tag.text
    names.append((user_name, full_name))
print(names)

````

#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code

html = requests.get(url).content

from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")

img_tag = ['img']

image_tags = [image for image in soup.find_all(img_tag)]
image_tags
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code

html = requests.get(url).content

from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")
````

````python
table_tag = ['table']

table = [table for table in soup.find_all(table_tag)][3]
````

````python
rows = table.find_all('tr')
rows = [row.text.strip().split("/n") for row in rows]
rows
````

````python
colnames = rows[1]
data = rows[5:]

df = pd.DataFrame(data, columns=colnames)
df
````

````python
# CORRECTION GUIDEE

# APPEL DE L'URL
res = requests.get(url)

# APPEL DE LA LIBRAIRIE BEAUTIFUL SOUP
soup = BeautifulSoup(res.content, 'html.parser')

# CREATION DE LA DATAFRAME
earthquakes = pd.DataFrame()

# CREATION D'UN ITERATEUR
i = 0

# BOUCLE SUR LA SECTION TBODY DE LA PAGE HTML

for tag in soup.find("tbody", {"id": "tbody"}).findAll("tr"): # RECHERCHE DES ID DANS TBODY ET DES ELEMENTS TR
    datetime = tag.find("td", {"class": "tabev6"}).find("a").text # RECHERCHE DES CLASSES DES ELEMENTS REQUIS
    lat_metric, lon_metric = [t.text for t in tag.findAll("td", {"class": "tabev1"})]
    lat_dir, lon_dir = [t.text for t in tag.findAll("td", {"class": "tabev2"})[:2]]
    region_tag = tag.find("td", {"class": "tb_region"}).text.strip()
    earthquakes = earthquakes.append(pd.DataFrame({"date": datetime.split()[0], 
                                                   "time": datetime.split()[1],
                                                   "lat_metric": lat_metric, 
                                                   "lat_dir": lat_dir,
                                                   "lon_metric": lon_metric,
                                                   "lon_dir": lon_dir, 
                                                   "region": region_tag}, index=[i]))
    i += 1
earthquakes
````
